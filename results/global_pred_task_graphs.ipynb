```python
import json
from matplotlib import pyplot as plt
```

```python
path = "./global_pred_microsoft_codebert-base.json"
with open(path) as f:
    data = json.load(f)
```

```python
for f in data:
    x = []
    y = []
    for l in data[f]:
        x.append(l)
        y.append(data[f][l]['accuracy'])
    plt.plot(x, y, label=f)
plt.legend()
# plt.show()
```
